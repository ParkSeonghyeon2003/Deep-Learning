```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터셋 준비
project_dir = '../../../'
df = pd.read_csv(project_dir+'data/BostonHousing.csv')
print(df.columns)
print(df.shape)
df_X = df[['crim', 'zn', 'indus', 'nox', 'rm', 'age',
    'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
print(df_X.shape)
df_y = df['medv']
print(df_y.shape)
```

```python
# 훈련/테스트 데이터 분리
train_X, test_X, train_y, test_y = \
    train_test_split(df_X, df_y, test_size=0.3,
                     random_state=1234)

# 훈련 모델 정의
model = LinearRegression()

# 훈련셋으로 훈련
model.fit(train_X, train_y)

# 테스트셋으로 예측
pred_y = model.predict(test_X)
print(pred_y)
```

```python
# coefficients & Intercept
print('Coefficients: {0:.2f},{1:.2f},{2:.2f} Intercept {3:.3f}'\
      .format(model.coef_[0], model.coef_[1], model.coef_[2],\
              model.intercept_))

# mean squared error
print('Mean squared error: {0:.2f}'.\
      format(mean_squared_error(test_y, pred_y)))

# r2 score
print('R2 score: {0:.2f}'.\
      format(r2_score(test_y, pred_y)))
```
